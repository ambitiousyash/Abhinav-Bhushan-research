```python
import pandas as pd

# Reading the specific sheets
metadata_df = pd.read_excel('/content/Supplementary Table S1.xlsx', sheet_name='IBD_metadata')
genes_df = pd.read_excel('/content/Supplementary Tables S12-S17.xlsx' , sheet_name='S14')

# Extracting all gene names from the first column
selected_genes = genes_df.iloc[:, 0].tolist()

# Transpose genes dataframe to align participant IDs correctly
genes_df_t = genes_df.set_index('Unnamed: 0').transpose().reset_index()
genes_df_t.rename(columns={'index': 'Participant_ID'}, inplace=True)

# Merge the two datasets based on Participant_ID
merged_df = pd.merge(metadata_df, genes_df_t, on='Participant_ID', how='inner')

# Output the Diagnosis column along with all genes
output_df = merged_df[['Diagnosis'] + selected_genes]

# Display the resulting dataframe
print(output_df.head(20))
# Display summary statistics for the genes
summary_stats = output_df.describe(include='all')
print(summary_stats)
```
