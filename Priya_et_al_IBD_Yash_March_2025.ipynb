```python
import pandas as pd

# Define file paths
file1_path = "/Supplementary Table S1.xlsx"
file2_path = "/Supplementary_Tables_S12_S17.xlsx"

# Load sheet names
xls1 = pd.ExcelFile(file1_path)
xls2 = pd.ExcelFile(file2_path)

# Display available sheet names in both files
xls1.sheet_names, xls2.sheet_names
```

```python
# Load the IBD_metadata sheet
ibd_metadata_df = pd.read_excel(xls1, sheet_name="IBD_metadata")

# Display the first few rows to understand its structure
ibd_metadata_df.head(20)
```

```python
# Extract relevant participant IDs
participant_ids = ibd_metadata_df["Participant_ID"].tolist()

# Load the S14 sheet
s14_df = pd.read_excel(xls2, sheet_name="S14")

# Display the first few rows to understand its structure
s14_df.head()
```

```python
# Transpose the S14 data to align participants as rows instead of columns
s14_df_transposed = s14_df.set_index(s14_df.columns[0]).T

# Rename index to match participant IDs
s14_df_transposed.index.name = "Participant_ID"
s14_df_transposed.reset_index(inplace=True)

# Merge with IBD metadata to include diagnosis information
merged_df = pd.merge(ibd_metadata_df, s14_df_transposed, on="Participant_ID", how="inner")

# Rearrange based on Diagnosis
sorted_df = merged_df.sort_values(by="Diagnosis")

# Display the first few rows of the sorted dataset
sorted_df.head()
```

```python
# Get summary statistics for each diagnosis group
summary_stats = sorted_df.groupby("Diagnosis").describe()
summary_stats
```
